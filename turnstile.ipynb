```python
station_groups = df.groupby(['CA', 'UNIT', 'SCP', 'STATION', 'LINENAME'])
```

```python
# station_groups.groups.keys()
```

```python
new_df = pd.DataFrame()
for name, group in station_groups:
#     print(name)
    group["ENTRY_INTERVAL"] = group["ENTRIES"].shift(-1) - group["ENTRIES"]
    group["EXIT_INTERVAL"] = group["EXITS"].shift(-1) - group["EXITS"]
    new_df = pd.concat([group, new_df])
```

```python
new_df.head()
```

```python
new_df['DATE_TIME'] = new_df["DATE"] + ' ' + new_df["TIME"]
```

```python
type(new_df['DATE_TIME'][0])
```

```python
agg_df = new_df.groupby(['STATION', 'LINENAME', 'DATE_TIME'])
```

```python
# agg_df.agg(['ENTRY_INTERVAL', 'EXIT_INTERVAL'])
final_data = agg_df[['ENTRY_INTERVAL', 'EXIT_INTERVAL']].sum().reset_index()
```

```python
# final_data.to_csv('data.csv')
# final_data.to_json('data.json', orient='records')
```

```python
final_data['DATE_TIME'] = final_data.DATE_TIME.astype(str)
```

```python
type(final_data['DATE_TIME'][0])
```

```python
plt.figure(figsize=(30,50))
plt.plot(final_data['DATE_TIME'], final_data['ENTRY_INTERVAL'],'o')
plt.plot(final_data['DATE_TIME'], final_data['EXIT_INTERVAL'],'o')
```

```python
final_data.describe()
```

```python
DATE_TIME = final_data.DATE_TIME.unique()

ord_by_date = {}

for date in DATE_TIME:
#     print(date)
    ord_by_date[f'{date}'] = final_data.query(f'DATE_TIME=="{date}"').to_dict('records')
```
