```python
KINGSTON_R124_R624_ONE = KINGSTON.query('UNIT=="R124" and CA=="R624" and SCP=="00-00-00"') 
```

```python
KINGSTON_R124_R624_ONE
```

```python
KINGSTON_R124_R624_ONE["ENTRY_INTERVAL"] = 0
KINGSTON_R124_R624_ONE["EXIT_INTERVAL"] = 0
```

```python
KINGSTON_R124_R624_ONE["ENTRY_INTERVAL"] = KINGSTON_R124_R624_ONE["ENTRIES"].shift(-1) - KINGSTON_R124_R624_ONE["ENTRIES"]
KINGSTON_R124_R624_ONE["EXIT_INTERVAL"] = KINGSTON_R124_R624_ONE["EXITS"].shift(-1) - KINGSTON_R124_R624_ONE["EXITS"]
```

```python
KINGSTON_R124_R624_ONE.head()
```

```python
KINGSTON_R124_R624_ONE["DATE_TIME"] = pd.to_datetime(KINGSTON_R124_R624_ONE["DATE"] + ' ' + KINGSTON_R124_R624_ONE["TIME"])
```

```python
plt.figure(figsize=(20,10))
plt.plot(KINGSTON_R124_R624_ONE['DATE_TIME'], KINGSTON_R124_R624_ONE['ENTRY_INTERVAL'])
plt.plot(KINGSTON_R124_R624_ONE['DATE_TIME'], KINGSTON_R124_R624_ONE['EXIT_INTERVAL'])
```

```python
station_groups = df.groupby(['CA', 'UNIT', 'SCP', 'STATION', 'LINENAME'])
```

```python
station_groups.groups.keys()
```

```python
new_df = pd.DataFrame()
for name, group in station_groups:
    print(name)
    group["ENTRY_INTERVAL"] = group["ENTRIES"].shift(-1) - group["ENTRIES"]
    group["EXIT_INTERVAL"] = group["EXITS"].shift(-1) - group["EXITS"]
    new_df = pd.concat([group, new_df])
```

```python
new_df.head()
```

```python
new_df['DATE_TIME'] = pd.to_datetime(new_df["DATE"] + ' ' + new_df["TIME"])
```

```python
new_df.head()
```

```python
agg_df = new_df.groupby(['STATION', 'LINENAME', 'DATE_TIME'])
```
